```python
import pandas as pd #   pandas


df = pd.read_json('data/recipes.json') #   DataFrame,    recipes.json
display(df.head()) #       DataFrame

display(df.shape[1])

all_ingredients = set()

for ing_list in df['ingredients']:
    for ing in ing_list:
        all_ingredients.add(ing)

display(len(all_ingredients))

def contains(ingredient_list, ingredient_name): #      
    return ingredient_name in ingredient_list


for ing in all_ingredients:
    df[ing] = df['ingredients'].apply(lambda ing_list: contains(ing_list, ing))

df['ingredients'] = df['ingredients'].apply(len) #         
display(df) #    DataFrame  
```

```python
df.to_csv('data/recipes.csv', index = False)
```
