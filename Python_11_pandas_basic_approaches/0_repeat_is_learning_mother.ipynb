Задание 0.5

Представьте ситуацию.
Вы работаете аналитиком некоторого YouTube-канала. У вас есть информация о просматриваемости роликов с этого канала в виде некоторого DataFrame со столбцами: идентификатор видео, количество просмотров, количество лайков, количество дизлайков и количество комментариев.

Продюсер канала просит вас показать ему основные статистические параметры всех признаков, описывающих ролики.

Какой из нижеперечисленных вариантов кода поможет вам вывести эту информацию?

- info()
- describe()
- describe(include=['object'])
- value_counts()

Ответ: describe()
